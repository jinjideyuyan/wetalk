# # 环视功能

```python
import re
 
phone = "2004959559 # 这是一个国外电话号码"
num = re.sub(r'(\d)((\d\d\d)+\b)', "\g<1>,\g<2>", phone)
print num
```

```python
import re
 
phone = "2004959559 # 这是一个国外电话号码"
num = re.sub(r'(?<=\d)(?=(\d\d\d)+\b)', ",", phone)
print num
```

```python
import re
 
line = "fsafsdjas\n\n\n\t \nffsdfsdfsd"
matchObj = re.findall(r'^.*$', line, re.M)
print matchObj
```

```python
import re
 
line = "fsafsdjas\n\n\n\t \nffsdfsdfsd"
newline = re.sub(r'^[ \t\r]*$', "<p>", line, 0, re.M)
print newline
```

```python
import re
 
line = "fsafsdjas\n\n\n\t \nffsdfsdfsd"
newline = re.sub(r'^\s*$', "<p>", line, 0, re.M)
print newline
```

```python
import re
 
line = "fsafsd@asdf"
matchObj = re.findall(r'\w+@\w+', line, re.M)
print matchObj
```

```python
### 高亮
import re
 
line = "fsafsdja"
newline = re.sub(r'^(.*)$', "\g<1>, \g<2>", line, 0, re.M)
print newline
```

```python
import re
 
line1 = "好的"
line2 = u"好的"

matchObj = re.findall(r'.', line1)
print matchObj
matchObj = re.findall(r'.', line2)
print matchObj
```

```python
### . 不匹配换行符
import re
 
line = "fsafsdjas\n\n\n\t \nffsdfsdfsd"

matchObj = re.findall(r'.+', line)
print matchObj

matchObj = re.findall(r'[^"]+', line)
print matchObj
```

```python
import re
 
line = "(((( Lots of punctuation )))"

matchObj = re.findall(r'(?P<word>\b\w+\b)', line)
print matchObj
```
